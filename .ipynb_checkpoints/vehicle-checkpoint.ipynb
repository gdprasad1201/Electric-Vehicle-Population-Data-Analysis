```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import re
import os
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

Electric_Vehicle_Pop_Data = 0

if not os.path.exists('Electric_Vehicle_Population_Data_Cleaned.csv'):

    Electric_Vehicle_Pop_Data = pd.read_csv('Electric_Vehicle_Population_Data.csv').dropna()

    Electric_Vehicle_Pop_Data['Vehicle Location'] = Electric_Vehicle_Pop_Data['Vehicle Location'].str.replace('POINT ', '')
    Electric_Vehicle_Pop_Data['Vehicle Location'] = Electric_Vehicle_Pop_Data['Vehicle Location'].apply(lambda x: re.split('[()]', x)[1])
    Electric_Vehicle_Pop_Data['Vehicle Location'] = Electric_Vehicle_Pop_Data['Vehicle Location'].str.replace(' ', ', ')

    Electric_Vehicle_Pop_Data[['Vehicle Longitude', 'Vehicle Latitude']] = Electric_Vehicle_Pop_Data['Vehicle Location'].str.split(',', expand=True)
    Electric_Vehicle_Pop_Data['Vehicle Longitude'] = Electric_Vehicle_Pop_Data['Vehicle Longitude'].astype(float)
    Electric_Vehicle_Pop_Data['Vehicle Latitude'] = Electric_Vehicle_Pop_Data['Vehicle Latitude'].astype(float)
    Electric_Vehicle_Pop_Data = Electric_Vehicle_Pop_Data.drop(columns = ['Vehicle Location'])

    Electric_Vehicle_Pop_Data['Model Year'] = Electric_Vehicle_Pop_Data['Model Year'].fillna(0).astype(int)
    Electric_Vehicle_Pop_Data['Postal Code'] = Electric_Vehicle_Pop_Data['Postal Code'].fillna(0).astype(int)
    Electric_Vehicle_Pop_Data['Legislative District'] = Electric_Vehicle_Pop_Data['Legislative District'].fillna(0).astype(int)
    Electric_Vehicle_Pop_Data['DOL Vehicle ID'] = Electric_Vehicle_Pop_Data['DOL Vehicle ID'].fillna(0).astype(int)
    Electric_Vehicle_Pop_Data['Base MSRP'] = Electric_Vehicle_Pop_Data['Base MSRP'].fillna(0).astype(float)
    Electric_Vehicle_Pop_Data['Electric Range'] = Electric_Vehicle_Pop_Data['Electric Range'].fillna(0).astype(float)

    # Write changes to a new csv file
    Electric_Vehicle_Pop_Data.to_csv('Electric_Vehicle_Population_Data_Cleaned.csv', index=False)

    Electric_Vehicle_Pop_Data.head()

else:
    Electric_Vehicle_Pop_Data = pd.read_csv('Electric_Vehicle_Population_Data_Cleaned.csv')
    
Electric_Vehicle_Pop_Data.head()
```

## What is the distribution of different types of electric vehicles (Battery Electric Vehicle vs. Plug-in Hybrid Electric Vehicle)?

```python
# What is the distribution of different types of electric vehicles (Battery Electric Vehicle vs. Plug-in Hybrid Electric Vehicle)?
data = Electric_Vehicle_Pop_Data['Electric Vehicle Type'].value_counts()
plt.figure(figsize=(15, 5))
sns.barplot(x = data.index, y = data.values)
plt.title('Distribution of Different Types of Electric Vehicles')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# What is the geographical distribution of electric vehicles across different counties?
data = Electric_Vehicle_Pop_Data['County'].value_counts()
plt.figure(figsize=(15, 5))
sns.barplot(x = data.index, y = data.values)
plt.title('Geographical Distribution of Electric Vehicles Across Different Counties')
plt.xlabel('County')

plt.xticks(rotation=90)
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# Which manufacturers have the highest number of electric vehicles registered?
data = Electric_Vehicle_Pop_Data['Make'].value_counts()
plt.figure(figsize=(15, 5))
sns.barplot(x = data.index, y = data.values)
plt.title('Manufacturers with the Highest Number of Electric Vehicles Registered')
plt.xlabel('Manufacturer')
plt.xticks(rotation=90)
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# What is the average electric range of vehicles by manufacturer and model?
data = Electric_Vehicle_Pop_Data.groupby(['Make', 'Model'])['Electric Range'].mean().sort_values(ascending=False).reset_index(name = 'Average Electric Range')
plt.figure(figsize=(15, 5))
sns.barplot(x = data['Make'] + ' ' + data['Model'], y = data['Average Electric Range'])
plt.title('Average Electric Range of Vehicles by Manufacturer and Model')
plt.xlabel('Manufacturer and Model')
plt.xticks(rotation=90)
plt.ylabel('Average Electric Range')
plt.show()
```
